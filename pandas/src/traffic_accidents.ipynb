```python
import matplotlib.pyplot as plt
latMean = df['LATITUDE'].mean()
lonMean = df['LONGITUDE'].mean()
#found mean of lattitude and longitude data
df['LATITUDE'].dropna(inplace=True)
df['LONGITUDE'].dropna(inplace=True)
#replace NAN with means
df['LATITUDE'].replace(0,latMean,inplace=True)
df['LONGITUDE'].replace(0,lonMean,inplace=True)
#replace 0 values with means since nyc is not at lat 0 or lon 0
#cleaned up data
boroughColors = {
    'BRONX': 'orange',         
    'BROOKLYN': 'blue',      
    'MANHATTAN': 'green',    
    'QUEENS': 'red',       
    'STATEN ISLAND': 'purple'  
}
#dict of colors for boroughs
for borough,color in boroughColors.items():
    boroughData = df[df['BOROUGH']==borough]
    plt.scatter(boroughData['LONGITUDE'],boroughData['LATITUDE'],c=color,label = borough)
plt.show()

```

note that dots that seem out of place were likely accidents with a longitude or lattitude of 0

## 10. which month did the most number of accidents occur?

```python
import calendar
df['CRASH DATE']= pd.to_datetime(df['CRASH DATE'])
#use datetime to convert format to date time format
df['MONTH'] = df['CRASH DATE'].dt.month
#extract month
accidentsPerMonth = df['MONTH'].value_counts()
print(accidentsPerMonth)
```

The month with the most accidents was January. I can conclude that accidents are more common in the winter months and much less common in spring and early summer. This could be attributed to worse road conditions in the winter and better road conditions in spring and early summer. July and August also have high number of accidents and this could be because this is peak travel season so more cars are simply on the road.
